```python
df = pd.read_csv('craigslist_detail_data.csv')
```

```python
df.tail(10)
```

```python
# first we clean the data to check duplicates, data types, and cases
df['city'].unique()
```

```python
# ugh the city names are such a hot mess, let's clean it step by step
# first get the upper and lower cases even
df["city"] = df["city"].str.title()
df["city"].unique()
```

```python
# build a replacement dict to replace all the messy cities
replacement = {
    'Don Mills': "North York",
    'Lawrence/Don Mills': 'North York',
    'Niagara Toronto C01 Bathurst/Lakeshore' : 'Toronto',
    'City Of Toronto': 'Toronto',
    'Little Portugal - Queen/Dovercourt': 'Toronto',
    'Westminster-Branson, Toronto': 'Toronto',
    'Banbury-Don Mills': 'Toronto',
    'West Hill': 'Scarborough',
    'Danforth Village': 'Toronto',
    'Bendale': 'Scarborough',
    'Uxbridge On': 'York Region',
    'Brock Area, On': 'York Region',
    'Waterfront Communities C1': 'Toronto',
    '1 Hurontario Street #2002': 'Mississauga',
    'Fletchers Meadow': 'Brampton',
    'Patterson, Vaughan': 'Vaughan',
    'Markham Village': 'Markham',
    'Highland Creek': 'Scarborough',
    'Willmott Neighbourhood': 'Milton',
    'Oakville-Milton': 'Oakville',
    'Old Meadowvale Village': 'Mississauga',
    'Eringate-Centennial-West Deane': 'Etobicoke',
    'Erin, Ontario': 'Erin',
    'Countryside & Dixie': 'Brampton',
    'Pharmacy And Sheppard': 'Scarborough',
    'Upper Beaches': 'Toronto',
    'Glen Williams': 'Halton Hills',
    'Downtown Toronto': 'Toronto',
    'Cooksville, Mississauga': 'Mississauga',
    'Rockwood Village':  'Etobicoke'
}
df["city"] = df['city'].replace(replacement)
```

```python
df['city'].unique()
```

```python
# since 'Edmonton','St. Petersburg, FL, USA', 'Salt Spring Island' is not Ontario, drop them
drop_rows = ['Edmonton','St. Petersburg, Fl, Usa', 'Salt Spring Island']
df = df[~df['city'].isin(drop_rows)]
```

```python
df['city'].unique()
```

```python
# check data types and convert dtypes if it doesn't make sense
df.info()
```

```python
df.head(2)
```

```python
# drop where beds and baths don't have data
df = df.dropna(subset = ['price', 'bedrooms', 'bathrooms', 'sqft'])
```

```python
df.head(20)
```

```python
# add pirce of $$$ per sqft
df['price_per_sqft'] = df.apply(
    lambda row: row['price']/row['sqft']
    if pd.notnull(row['price']) and pd.notnull(row['sqft'])
    else None, # regular expression not regular if else
    axis = 1 # tell the lamda goes to all the rows instead of columns
)
```

```python
df.head(10)
```

```python
# round price per sqft
df['price_per_sqft'] = round(df['price_per_sqft'], 2)
```

```python
df.to_csv("craigslist_analysis_data(clean).csv", index=False)
```
